#Funcion Recomendar

```python
def Recomendador(Unser_ID):
  #Esta funcion recibe un usuario id, lo busca en la base de datos y si lo encuentra le da una recomendacion de 40 contenidos recomendados,
  #caso contrario recomienda los 20 contenidos mas populares.
  
  if Unser_ID in list(interaction.index):
    
    user_x = user_dict[Unser_ID]
    n_users, n_items = interaction.shape
    item_ids = np.arange(n_items)
    preds = model.predict(user_ids=user_x, item_ids = item_ids)

    scores = pd.Series(preds)
    scores.index = interaction.columns
    Recomendadas = list(pd.Series(scores.sort_values(ascending=False).index))[:20] #obtenemos las primeras 10

    #chequeamos que el contenido no fue previamente visto
    vio=db_train[db_train['account_id']==2].content_id.unique()
    for i in vio:
      for j in Recomendadas:
        if i==j:
          Recomendadas.remove(i)


  else:
    Recomendadas=populares

  return Recomendadas
```

```python
tabla_recomendada = pd.DataFrame(tabla_recomendada)
```

```python
aps = [] # lista vacía para ir almacenando la AP de cada recomendación


for pred, label in tabla_recomendada[["content_id", "Popular"]].values:
  n = len(pred) # cantidad de elementos recomendados
  arange = np.arange(n, dtype=np.int32) + 1. # indexamos en base 1 
  rel_k = np.in1d(pred[:n], label) # lista de booleanos que indican la relevancia de cada ítem
  tp = np.ones(rel_k.sum(), dtype=np.int32).cumsum() # lista con el contador de verdaderos positivos
  denom = arange[rel_k] # posiciones donde se encuentran los ítems relantes
  ap = (tp / denom).sum() / len(label) # average precision
  aps.append(ap)
```

```python
MAP = np.mean(aps)
print(f'mean average precision = {round(MAP, 5)}')
```
